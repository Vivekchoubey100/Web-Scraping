```python
df = pd.read_csv("link_by_areas.csv")
```

```python
list_of_job_type = df.type.values.tolist()
```

```python
list_of_url = df.link.values.tolist()
#list_of_url
```

```python
driver = webdriver.Chrome(path)
```

```python
count = 0
type =0
```

```python
for l in list_of_url:
    print(l)
    driver.get(l)
    time.sleep(5)
    
    def append_dict_as_row(file_name,list_of_elem,fiel_names):
        with open(file_name, 'a+', newline='') as write_obj:
            dict_writer = DictWriter(write_obj, fieldnames=field_names)
            dict_writer.writerow(dict_of_elem)
    
    while(1):
        driver.execute_script('window.scrollTo(0, 5000);')
        time.sleep(5)
        content = driver.page_source
        soup = BeautifulSoup(content,'html.parser')
        
        Total_jobs = (soup.find_all('article',attrs={'class':'jobTuple bgWhite br4 mb-8'}))
        
        links_of_jobs = []
        
        field_names = ['Type','Url','Title','Company Name','Experience','Salary','Location','Description','Tags Associated','Post Date','Current Date']
        
        for i in Total_jobs:
            count = count+1
            
            title = (i.find('a', attrs={'class':'title fw500 ellipsis'})).text
            company_name = (i.find('a', attrs={'class':'subTitle ellipsis fleft'})).text
            
            try:
                experience = (i.find('li', attrs={'class':'fleft grey-text br2 placeHolderLi experience'})).text
            except:
                experience = 'NA'
                
            salary = (i.find('li', attrs={'class':'fleft grey-text br2 placeHolderLi salary'})).text
            location = (i.find('li', attrs={'class':'fleft grey-text br2 placeHolderLi location'})).text
            description = ((i.find('div', attrs={'class':'job-description fs12 grey-text'})).text).encode("utf-8")
            
            try:
                tag_associated = (i.find('ul', attrs={'class':'tags has-description'})).findChildren()
                tag_list = []
                for m in tag_associated:
                    txt = m.text
                    tag_list.append(txt)
            except:
                tag_associated = 'NA'
                
            try:
                posted_date = (i.find('div', attrs={'class':'type br2 fleft grey'})).text
                dstr_to_int = re.findall(r'\d+', posted_date)
                dint = int(dstr_to_int[0])
                if(dint<30):
                    dt = datetime.date.today() - datetime.timedelta(dint)
                else:
                    dt = posted_date
            except:
                try:
                    posted_date = (i.find('div', attrs={'class':'type br2 fleft green'})).text
                    try:
                        dstr_to_int = re.findall(r'\d+', posted_date)
                        dint = int(dstr_to_int[0])
                        if(dint<30):
                            dt = datetime.date.today() - datetime.timedelta(dint)
                        else:
                            dt = posted_date
                    except:
                        dt = datetime.date.today()
                except:
                        dt = 'NA'

                    
                  
            
            print(title)
            print(company_name)
            print(experience)
            print(salary)
            print(location)
            print(description)
            
            print(tag_list)
            print(dt)
            print(count)
            current_date = datetime.date.today()
            dict_of_elem = {'Type':list_of_job_type[type],'Url':l,'Title':title,'Company Name':company_name,'Experience':experience,'Salary':salary,'Location':location,'Description':description,'Tags Associated':tag_list,'Post Date':dt,'Current Date':current_date}
            append_dict_as_row('Naukri.csv', dict_of_elem, field_names)
            print('------------------------------------------------------------------')
        break
    print(type)    
    type += 1
```
